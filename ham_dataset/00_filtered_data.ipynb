# HAM10000 Metadata Filtering

Filtrimi i metadata folder dhe mbajtja vetem e imazheve qe ekzistojne ne folderat e ri per klasifikim binar (mel dhe nv)

<h4> Dataset Paths <h4>

```python
import os
import pandas as pd

BASE_DIR = os.getcwd()

METADATA_PATH = os.path.join(
    BASE_DIR,
    "HAM10000_metadata.csv"
)

MEL_DIR = os.path.join(
    BASE_DIR,
    "mel"
)

NV_DIR = os.path.join(
    BASE_DIR,
    "nv"
)

#print("Base directory:", BASE_DIR)
```

<h4> Funksioni per leximin e imazheve <h4>

```python
def get_image_ids(folder_path):
    ids = set()

    for fname in os.listdir(folder_path):
        if fname.lower().endswith(".jpg"):
            ids.add(os.path.splitext(fname)[0])

    return ids
```

<h4> Leximi i metadata <h4>

```python
df = pd.read_csv(METADATA_PATH)

print("Columns:")
print(df.columns.tolist())

print("\nTotal rows:")
print(len(df))

df.head()
```

<h4> Gjetja automatike e kolones ID <h4>

```python
possible_cols = [
    "image_id",
    "image",
    "lesion_id"
]

id_col = None

for col in possible_cols:
    if col in df.columns:
        id_col = col
        break

if id_col is None:
    raise ValueError(
        "Could not find an image identifier column in metadata."
    )

#print("Using column:", id_col)
```

<h4> Leximi i fotove nga folderat <h4>

```python
mel_ids = get_image_ids(MEL_DIR)
nv_ids = get_image_ids(NV_DIR)

valid_ids = mel_ids.union(nv_ids)

print("Mel images:", len(mel_ids))
print("NV images:", len(nv_ids))
print("Total images:", len(valid_ids))
```

<h4> Filtrimi i Metadata <h4>

```python
filtered_df = df[
    df[id_col].isin(valid_ids)
].reset_index(drop=True)

print(
    f"Rows kept: {len(filtered_df)} / {len(df)}"
)

filtered_df.head()
```

<h4> Ruajtja e CSV-se se filtruar <h4>

```python
output_path = os.path.join(
    BASE_DIR,
    "HAM10000_metadata_filtered.csv"
)

filtered_df.to_csv(
    output_path,
    index=False
)

#print("Saved to:")
#print(output_path)
```

<h4> Verifikimi <h4>

```python
saved_df = pd.read_csv(output_path)

print("Saved rows:", len(saved_df))
saved_df.head()
```
